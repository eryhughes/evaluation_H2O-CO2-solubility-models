## Precision and accuracy

```python
# import data
ShishkinaIdealMixing = pd.read_csv("results_pvsat_MC_VESIcal/results_pvsat_MC_VESIcal_ShishkinaIdealMixing.csv")
Dixon = pd.read_csv("results_pvsat_MC_VESIcal/results_pvsat_MC_VESIcal_Dixon.csv")
IaconoMarziano = pd.read_csv("results_pvsat_MC_VESIcal/results_pvsat_MC_VESIcal_IaconoMarziano.csv")
IaconoMarziano = IaconoMarziano[IaconoMarziano['P_bar sd model']<=1000.]
MagmaSat = pd.read_csv("results_pvsat_MC_VESIcal/results_pvsat_MC_VESIcal_MagmaSat.csv")
Liu = pd.read_csv("results_pvsat_MC_VESIcal/results_pvsat_MC_VESIcal_Liu.csv")
```

### Pressure of vapor saturation (Pvsat)

```python
for model in ['Dixon','ShishkinaIdealMixing','IaconoMarziano','Liu','MagmaSat']:
    dataset = eval(model)
    result1 = pd.DataFrame([[model]])
    for option in [1,2,3,4,5]:
        for calib in ['N','Y']:
            if option == 1:
                dataset1 = dataset
                name = 'all'
            elif option == 2:
                dataset1 = dataset[(dataset['Composition']=='Basalt') | (dataset['Composition']=='Basaltic-andesite')]
                name = 'BBA'
            elif option == 3:
                dataset1 = dataset[(dataset['Composition']=='Basanite') | (dataset['Composition']=='Foidite')| (dataset['Composition']=='Phonolite')| (dataset['Composition']=='Phonotephrite')| (dataset['Composition']=='Tephrite')| (dataset['Composition']=='Trachyandesite')| (dataset['Composition']=='Trachybasalt')| (dataset['Composition']=='Trachyte')]
                name = 'AR'
            elif option == 4:
                 dataset1 = dataset[(dataset['Composition']=='Andesite') | (dataset['Composition']=='Dacite')]
                 name = 'I'
            elif option == 5:
                dataset1 = dataset[(dataset['Composition']=='Rhyolite')]
                name ='R'
            data = dataset1[dataset1['calib_'+model]==calib]
            number = data['P_bar model'].count()
            hist = ((data['P_bar model']-data['P_bar'])/data['P_bar'])
            accuracy = hist.mean()
            precision = hist.std()
            result2 = pd.DataFrame([[number,accuracy,precision]])
            result1 = pd.concat([result1,result2],axis=1)
        if option == 5:
            result3 = result1
    if model =='Dixon':
        result4 = result3
    else:
        result4 = pd.concat([result4,result3])
result4.to_csv("n_av_std_pvsat.csv", index=False, header=False)
```

### H2O mole fraction in the vapor (xvH2O)

```python
for model in ['Dixon','ShishkinaIdealMixing','IaconoMarziano','Liu','MagmaSat']:
    dataset = eval(model)
    result1 = pd.DataFrame([[model]])
    for option in [1,2,3,4,5]:
        for calib in ['N','Y']:
            if option == 1:
                dataset1 = dataset
                name = 'all'
            elif option == 2:
                dataset1 = dataset[(dataset['Composition']=='Basalt') | (dataset['Composition']=='Basaltic-andesite')]
                name = 'BBA'
            elif option == 3:
                dataset1 = dataset[(dataset['Composition']=='Basanite') | (dataset['Composition']=='Foidite')| (dataset['Composition']=='Phonolite')| (dataset['Composition']=='Phonotephrite')| (dataset['Composition']=='Tephrite')| (dataset['Composition']=='Trachyandesite')| (dataset['Composition']=='Trachybasalt')| (dataset['Composition']=='Trachyte')]
                name = 'AR'
            elif option == 4:
                 dataset1 = dataset[(dataset['Composition']=='Andesite') | (dataset['Composition']=='Dacite')]
                 name = 'I'
            elif option == 5:
                dataset1 = dataset[(dataset['Composition']=='Rhyolite')]
                name ='R'
            data = dataset1[dataset1['calib_'+model]==calib]
            number = data['XH2Of'].count()
            hist = ((data['xgH2O model']-data['XH2Of']))
            accuracy = hist.mean()
            precision = hist.std()
            result2 = pd.DataFrame([[number,accuracy,precision]])
            result1 = pd.concat([result1,result2],axis=1)
        if option == 5:
            result3 = result1
    if model =='Dixon':
        result4 = result3
    else:
        result4 = pd.concat([result4,result3])
result4.to_csv("n_av_std_xH2O.csv", index=False, header=False)
```

### Example calculation for an individual composition: Pavlof

Mangan, M. T., Sisson, T. W., Hankins, W. Ben, Shimizu, N., & Vennemann, T. (2021). Constraints on deep, CO2-rich degassing at arc volcanoes from solubility experiments on hydrous basaltic andesite of Pavlof Volcano, Alaska Peninsula, at 300 to 1200 MPa. American Mineralogist, 106(5), 762–773. https://doi.org/10.2138/am-2021-7531

```python
for model in ['Dixon','ShishkinaIdealMixing','IaconoMarziano','Liu','MagmaSat']:
    dataset = eval(model)
    result1 = pd.DataFrame([[model]])
    if model in ['Dixon','ShishkinaIdealMixing','IaconoMarziano','Liu']:
        data = dataset[(dataset['Location']=='Pavlof, Alaksa USA')]
    elif model in ['MagmaSat']:
        data = dataset[(dataset['Location']=='Pavlof, Alaska, USA')]
    number_pvsat = data['P_bar model'].count()
    hist_pvsat = ((data['P_bar model']-data['P_bar'])/data['P_bar'])
    accuracy_pvsat = hist_pvsat.mean()
    precision_pvsat = hist_pvsat.std()
    number_XH2O = data['XH2Of'].count()
    hist_XH2O = ((data['xgH2O model']-data['XH2Of']))
    accuracy_XH2O = hist_XH2O.mean()
    precision_XH2O = hist_XH2O.std()
    result2 = pd.DataFrame([[number_pvsat,accuracy_pvsat,precision_pvsat,number_XH2O,accuracy_XH2O,precision_XH2O]])
    result1 = pd.concat([result1,result2],axis=1)
    if model =='Dixon':
        result3 = result1
    else:
        result3 = pd.concat([result3,result1])
result3.to_csv("n_av_std_pvsat_XH2O.csv", index=False, header=False)
```
